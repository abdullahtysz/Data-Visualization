```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x="salary",y = "work experience",data = df);
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, hue="training level");
# hue  : Bu argümanın değerini 'training level' yaptığımız zaman traing leveline göre renklendirme yapıyor. 
# Bu özellik sayesinde grafiğimize üçüncü bir boyut katmış oluyoruz
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, hue="level of education", palette="Dark2");
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, hue="level of education")
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, size = "salary");
# size'ı salary yaptıgımızda yogunlugu maaşlara gore ve kalınlıklarına gore verir
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, s=200); 
# s= işaretlerin büyüklüğünü ayarlar
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, s=200,alpha= 0.2);
# alpha saydamlık katar
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, s=200, style="level of education");
# style ile biraz once hue ile renklendirdiğimiz 3. boyutu şekillerle vermiş olduk
```

```python
plt.figure(figsize=(12,4)) 
sns.scatterplot(x="salary",y = "work experience",data = df, s=200, style="level of education",hue="level of education");
# hem şekil hem renkler ile daha anlamlı bir grafik elde ettik
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(15,8))
sns.rugplot(x = "salary",data = df, height= 0.5);
# rugplot(), sadece tek değişkenli bir dağılım üzerindeki her nokta için çizgi çizer. 
```

```python
sns.displot(data= df, x = "salary",bins= 155);
```

```python
sns.set(style="darkgrid")
sns.displot(data= df, x = "salary",bins= 15);
```

```python
sns.set(style="darkgrid")
sns.displot(data= df, x = "salary",bins= 15,color="red",edgecolor="blue");
```

```python
sns.set(style="darkgrid")
sns.displot(data= df, x = "salary",bins= 15,color="red",edgecolor="blue",lw=4);
```

```python
sns.set(style="darkgrid")
sns.displot(data= df, x = "salary",bins= 15,color="red",edgecolor="blue",lw=4, ls = "--");
```

```python
sns.histplot(data= df, x = "salary",bins= 15);
```

```python
sns.kdeplot(data = df, x = "salary");
# KDE Plot (Kernel Density Estimate) sürekli bir değerin olasılık yoğunluklarını görselleştirmek adına kullanılır. 
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages 
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=["age"])
sample_ages
```

```python
sns.rugplot(data = sample_ages, x = "age");
```

```python
sns.displot(data = sample_ages, x = "age",rug=True)
```

```python
sns.displot(data = sample_ages, x = "age",rug=True,bins = 30);
```

```python
sns.displot(data = sample_ages, x = "age",rug=True,bins = 30, kde  = True);
# sürekli bir değerin olasılık yoğunluklarını görselleştirmek adına kullanılır.
```

```python
sns.kdeplot(data = sample_ages, x = "age");
```

```python
sns.kdeplot(data = sample_ages, x = "age", clip = [0,100]);
# clip , Eğrinin görüntüleneceği aralık
```

```python
sns.kdeplot(data = sample_ages, x = "age", clip = [0,100],bw_adjust=0.10);
#bw_adjust, argüman kde eğrisinin düzgünlük derecesini ayarlar, default değeri 1 dir 
```

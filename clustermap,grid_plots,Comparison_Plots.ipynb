```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(10,4),dpi = 200)
sns.boxplot(data = df , y  = "math score", x = "parental level of education",
            hue = "test preparation course")
plt.legend(bbox_to_anchor = (1.05,0.5))
```

```python
plt.figure(figsize=(10,4),dpi = 200)
sns.boxplot(data = df , x  = "reading score", y = "parental level of education",
            hue = "test preparation course")
plt.legend(bbox_to_anchor = (1.05,0.5))
```

```python
plt.figure(figsize=(10,8),dpi = 200)
sns.violinplot(data = df , x  = "reading score", y = "parental level of education",
            hue = "test preparation course",
            palette = "Set2",
            split = True )
plt.legend(bbox_to_anchor = (1.25,0.5))
```

```python
plt.figure(figsize=(10,8),dpi = 200)
sns.violinplot(data = df , x  = "reading score", y = "parental level of education",
            hue = "test preparation course",
            palette = "Set2",
            split = True,
            inner = "quartile" )
plt.legend(bbox_to_anchor = (1.25,0.5))
```

```python
sns.swarmplot(data = df , x = "math score");
```

```python
plt.figure(figsize=(8,5), dpi = 100)
sns.swarmplot(data = df , x = "math score",y = "gender",size = 2, hue = "test preparation course")
```
